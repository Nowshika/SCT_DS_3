```python
#BANK MARKETING ANALYSIS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.preprocessing import LabelEncoder, StandardScaler
from matplotlib.colors import LinearSegmentedColormap
import warnings
warnings.filterwarnings("ignore")

# Set up visualization style
plt.style.use('dark_background')
plt.rcParams['figure.facecolor'] = '#0a0a0a'
plt.rcParams['axes.facecolor'] = '#1a1a1a'
plt.rcParams['text.color'] = '#ffffff'
plt.rcParams['axes.labelcolor'] = '#ffffff'
plt.rcParams['xtick.color'] = '#cccccc'
plt.rcParams['ytick.color'] = '#cccccc'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['grid.color'] = '#333333'
plt.rcParams['grid.alpha'] = 0.4

# Custom color palette
custom_palette = ['#025043', '#036c5f', '#037c6e', '#048c7f', '#05998c', '#06a699', '#07b3a6', '#08c0b3', '#09cdc0', '#0adacd', '#0be7da', '#0cf4e7']
sns.set_palette(custom_palette)

print("="*80)
print("BANK MARKETING CAMPAIGN ANALYSIS")
print("="*80)

# Load data with error handling
file = 'bank-full.csv'
try:
    df = pd.read_csv(file, sep=';')
except FileNotFoundError:
    print(f"Error: File '{file}' not found.")
    exit(1)
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file. Check the separator or file format.")
    exit(1)

# Subsample for faster runtime (use 20% of data for prototyping)
df = df.sample(frac=0.2, random_state=42)
print(f"\nDataset Shape (Subsampled): {df.shape}")
print(f"Target Distribution:\n{df['y'].value_counts()}")
print(f"Target Distribution (%):\n{df['y'].value_counts(normalize=True) * 100}")

print("\nMissing Values:")
print(df.isnull().sum())

# Feature engineering
df['age_duration_interaction'] = df['age'] * df['duration']
df['total_contacts'] = df['campaign'] + df['previous']
df['age_binned'] = pd.cut(df['age'], bins=5, labels=False)
df['duration_binned'] = pd.cut(df['duration'], bins=5, labels=False)

# Encode categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

X = df.drop('y', axis=1)
y = df['y']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Define models
models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=3),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42, max_depth=10, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42, max_depth=5),
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Linear SVC': LinearSVC(random_state=42, max_iter=1000),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Naive Bayes': GaussianNB()
}

print("\n" + "="*80)
print("TRAINING MODELS")
print("="*80)

for name, model in models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train)

results = {}
predictions = {}

print("\n" + "="*80)
print("MODEL EVALUATION RESULTS")
print("="*80)

# Custom colormap for confusion matrix
custom_cmap = LinearSegmentedColormap.from_list("custom", ["#025043", "#05998c", "#0cf4e7"])

for name, model in models.items():
    print(f"\n{'-'*50}")
    print(f"MODEL: {name}")
    print(f"{'-'*50}")
    
    y_pred = model.predict(X_test)
    predictions[name] = y_pred
    
    # Handle ROC probabilities
    if hasattr(model, 'predict_proba'):
        y_proba = model.predict_proba(X_test)[:, 1]
    else:
        print(f"Warning: {name} does not support predict_proba; using decision function.")
        y_proba = model.decision_function(X_test)
        y_proba = (y_proba - y_proba.min()) / (y_proba.max() - y_proba.min())

    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    
    print(f"Accuracy: {acc:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
    
    cv_scores = cross_val_score(model, X_scaled, y, cv=5)
    print(f"Cross-validation Score: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=custom_cmap, 
                xticklabels=['No', 'Yes'], yticklabels=['No', 'Yes'],
                cbar_kws={'shrink': 0.8})
    plt.title(f"Confusion Matrix - {name}", fontsize=16, pad=20)
    plt.xlabel('Predicted', fontsize=12)
    plt.ylabel('Actual', fontsize=12)
    plt.tight_layout()
    plt.show()

    # ROC curve
    plt.figure(figsize=(8, 6))
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, color='#05998c', lw=3, label=f'ROC curve (AUC = {roc_auc:.3f})')
    plt.plot([0, 1], [0, 1], color='#037c6e', lw=2, linestyle='--', alpha=0.8)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title(f'ROC Curve - {name}', fontsize=16, pad=20)
    plt.legend(loc="lower right", fontsize=12)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    # Feature importance for tree-based models
    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        top_features = min(15, len(indices))
        
        plt.figure(figsize=(12, 8))
        colors = [custom_palette[i % len(custom_palette)] for i in range(top_features)]
        bars = plt.bar(range(top_features), importances[indices[:top_features]], color=colors)
        plt.xticks(range(top_features), [X.columns[i] for i in indices[:top_features]], rotation=45, ha='right')
        plt.title(f"Top {top_features} Feature Importances - {name}", fontsize=16, pad=20)
        plt.xlabel('Features', fontsize=12)
        plt.ylabel('Importance', fontsize=12)
        plt.tight_layout()
        plt.show()

# Optimized Decision Tree Visualization
print("\n" + "="*50)
print("DECISION TREE VISUALIZATION")
print("="*50)

simple_tree = DecisionTreeClassifier(random_state=42, max_depth=3, min_samples_leaf=50)
simple_tree.fit(X_train, y_train)

# Export tree as text for clarity
tree_rules = export_text(simple_tree, feature_names=list(X.columns), decimals=2)
print("Decision Tree Rules (Max Depth: 3, Min Samples: 50):")
print(tree_rules)

# Visualize tree with black text
plt.figure(figsize=(20, 10))
tree_plot = plot_tree(simple_tree, 
                      feature_names=X.columns, 
                      class_names=['No', 'Yes'], 
                      filled=True, 
                      rounded=True, 
                      fontsize=10,
                      proportion=True,
                      impurity=True,
                      node_ids=True,
                      precision=2)

# Customize text color to black
for text in tree_plot:
    text.set_color('black')

plt.title("Decision Tree (Max Depth: 3, Min Samples: 50)", fontsize=20, pad=30)
plt.tight_layout()
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
plt.show()

print("\n" + "="*80)
print("COMPREHENSIVE ANALYSIS SUMMARY")
print("="*80)

sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
print("\nModel Performance Ranking:")
for i, (name, score) in enumerate(sorted_results, 1):
    print(f"{i}. {name}: {score:.4f}")

best_model = sorted_results[0][0]
print(f"\nBest Performing Model: {best_model} (Accuracy: {sorted_results[0][1]:.4f})")

print("\n" + "="*80)
print("BUSINESS INSIGHTS")
print("="*80)

print("\nKey Findings:")
print("1. Model Performance Analysis:")
for name, score in sorted_results[:3]:
    print(f"    {name}: {score:.1%} accuracy")

print("\n2. Feature Importance Analysis:")
if hasattr(models['Random Forest'], 'feature_importances_'):
    rf_importances = models['Random Forest'].feature_importances_
    rf_indices = np.argsort(rf_importances)[::-1]
    print("   Top 5 Most Important Features (Random Forest):")
    for i in range(min(5, len(rf_indices))):
        feature_name = X.columns[rf_indices[i]]
        importance = rf_importances[rf_indices[i]]
        print(f"    {feature_name}: {importance:.3f}")

print("\n3. Business Recommendations:")
print("    Target customers with high duration and successful prior outcomes (poutcome).")
print("    Shorten calls for low-probability clients to reduce campaign costs.")
print("    Use Random Forest or Gradient Boosting for reliable predictions.")
print("    Implement A/B testing for high-importance features like duration.")

print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)
```
